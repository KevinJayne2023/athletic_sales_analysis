```python
# Import Libraries and Dependencies
import pandas as pd
import datetime 
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
csv2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
csv2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
csv2020
```

```python
# Display the 2021 sales DataFrame
csv2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
csv2020.dtypes
```

```python
# Check the 2021 sales data types.
csv2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([csv2020, csv2021], ignore_index=True)
combined_df.head()
```

```python
# Check if any values are null.
combined_df.isnull().sum()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_df = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
grouped_df.rename(columns= {'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
grouped_df.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_df = combined_df.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_sales = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

grouped_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
grouped_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_sales = combined_df.pivot_table(values='total_sales', index=['region', 'state', 'city'], aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_sales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_retailer = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()

grouped_retailer.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
grouped_retailer.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_retailer = combined_df.pivot_table(values='total_sales', index=['retailer', 'region', 'state', 'city'], aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_retailer.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_retailer.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_sales_data = combined_df[combined_df['product'] == "Women's Athletic Footwear"]

#Or
womens_sales_data_filter = combined_df.filter(like="Women's Athletic Footwear", axis=0)

womens_sales_data.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_womens = womens_sales_data.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

grouped_womens.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
grouped_womens.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_pivot_data = womens_sales_data.pivot_table(values='units_sold', index=['retailer', 'region', 'state', 'city']).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_pivot_data.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
womens_pivot_data.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_day = combined_df.pivot_table(values='total_sales', index='invoice_date', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_day.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
womens_day
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resampled_womens_day = womens_day.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_womens_day = resampled_womens_day.sort_values(by='Total Sales', ascending=False)

sorted_womens_day
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_week = womens_day.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_womens_week = womens_week.sort_values(by='Total Sales', ascending=False)
sorted_womens_week
```
